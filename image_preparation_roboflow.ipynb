```python
# function 1
# Function to count the number of image files in a directory
def count_images(directory):
    image_extensions = ['.jpg', '.jpeg', '.png', '.gif', '.bmp']  # Add more extensions if needed
    count = 0
    for file in os.listdir(directory):
        if any(file.lower().endswith(ext) for ext in image_extensions):
            count += 1
    return count

# Function to display folder names, number of images inside, number of folders, and total number of images
def display_folder_info(root_folder):
    num_folders = 0
    total_images = 0
    for folder_name in os.listdir(root_folder):
        folder_path = os.path.join(root_folder, folder_name)
        if os.path.isdir(folder_path):
            num_folders += 1
            num_images = count_images(folder_path)
            total_images += num_images
            print(f"Folder Name: {folder_name}, Number of Images: {num_images}")
    
    print(f"\nNumber of Folders: {num_folders}")
    print(f"Total Number of Images: {total_images}")
```

```python
# function 2
import os
import shutil

# Function to count the number of image files in a directory
def count_images(directory):
    image_extensions = ['.jpg', '.jpeg', '.png', '.gif', '.bmp']  # Add more extensions if needed
    count = 0
    for file in os.listdir(directory):
        if any(file.lower().endswith(ext) for ext in image_extensions):
            count += 1
    return count

# Function to create new folders with a maximum of max_images_per_subfolder images
def create_folders_sample(source_folder, dest_folder, max_images_per_subfolder):
    for folder_name in os.listdir(source_folder):
        folder_path = os.path.join(source_folder, folder_name)
        if os.path.isdir(folder_path):
            os.makedirs(os.path.join(dest_folder, folder_name), exist_ok=True)
            count = 0
            dest_subfolder_path = os.path.join(dest_folder, folder_name)
            os.makedirs(dest_subfolder_path, exist_ok=True)
            for file_name in os.listdir(folder_path):
                if count >= max_images_per_subfolder:
                    break
                
                src_file_path = os.path.join(folder_path, file_name)
                dest_file_path = os.path.join(dest_subfolder_path, file_name)
                shutil.copy(src_file_path, dest_file_path)
                count += 1
```

```python
source_folder = rf"{home}/garbage_classification"  # Source folder with images
dest_folder = rf"{home}/garbage_classification_100items_12class"  # Destination folder for organized images
max_images_per_folder = 100  # Maximum images per subfolder
    
create_folders_sample(source_folder, dest_folder, max_images_per_folder)
print("Folders created successfully with the specified image limit per folder.")
```

```python
# function 3
import os

def rename_images_in_subfolders(parent_folder):
    # Iterate through subfolders in the parent folder
    for folder_name in os.listdir(parent_folder):
        folder_path = os.path.join(parent_folder, folder_name)
        if os.path.isdir(folder_path):
            image_files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]
            # Rename image files in the subfolder
            for idx, image_name in enumerate(image_files):
                extension = os.path.splitext(image_name)[1]
                new_name = f"{folder_name}_{idx:05d}{extension}"  # Format the new name with leading zeros
                old_path = os.path.join(folder_path, image_name)
                new_path = os.path.join(folder_path, new_name)
                os.rename(old_path, new_path)
                print(f"Renamed: {image_name} -> {new_name}")

```

```python
parent_folder = rf"{home}/garbage_classification_100items_12class"  # Path to the parent folder containing subfolders
rename_images_in_subfolders(parent_folder)
print("Image renaming completed successfully.")
```

```python
#function 4
import os

def remove_subset_word(folder_path):
    # Get a list of all subfolders in the specified folder path
    subfolders = [f.name for f in os.scandir(folder_path) if f.is_dir()]

    # Iterate through each subfolder and rename if necessary
    for subfolder in subfolders:
        new_name = subfolder.replace('_', '')  # Remove the word 'subset' ### replace this with (text/symbol to replace) (what to replace with)
        if new_name != subfolder:  # Check if renaming is needed
            old_path = os.path.join(folder_path, subfolder)
            new_path = os.path.join(folder_path, new_name)
            os.rename(old_path, new_path)
            print(f'Renamed "{subfolder}" to "{new_name}"')
```
